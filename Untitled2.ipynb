```python
from pmdarima.arima import auto_arima
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
from math import sqrt
import numpy as np
import pandas as pd
import os
from matplotlib import pyplot as plt
path=os.getcwd()
file=open("attributes.csv",'a')
dateparse= lambda x:pd.datetime.strptime(x,'%d-%m-%Y')
df=pd.read_csv("sample.csv",parse_dates=['Date'],date_parser=dateparse)
df.head()
df1=pd.read_csv("test.csv",parse_dates=['Date'],date_parser=dateparse)
df1.head()
for i in range(1,260):
    x=str(i)
    aa=df[df.Productcat=="ProdCat"+x]
    l=aa.values
    train=l[:,1:3]
    bb=df1[df1.Productcat=="ProdCat"+x]
    l1=bb.values
    valid=l1[:,1:3]
    training=l[:,2]
    validation=l1[:,2]
    count=valid.shape
    model = auto_arima(training, start_p=1, start_q=1,max_p=3, max_q=3, m=12,start_P=0, seasonal=True,d=1, D=1, trace=True,error_action='ignore',suppress_warnings=True)
    model.fit(training)
    forecast = model.predict(n_periods=count[0])
    forecast = pd.DataFrame(forecast,index = bb.index,columns=['Prediction'])
    #rms=np.sqrt(np.mean(np.power((np.array(validation)-np.array(forecast['Prediction'])),2)))
    rmse=sqrt(mean_squared_error(validation,forecast['Prediction']))
    r2=r2_score(validation,forecast['Prediction'])
    mae=mean_absolute_error(validation,forecast['Prediction'])
    file.write(str("ProdCat19"))
    file.write(",")
    file.write(str(rmse))
    file.write(",")
    file.write(str(r2))
    file.write(",")
    file.write(str(mae))
    file.write("\n")
file.close()
#plt.plot(l[:,1],l[:,2])
#plt.plot(l1[:,1],l1[:,2])
#plt.plot(l1[:,1],forecast['Prediction'])
#plt.plot(forecast['Prediction'])
#plt.show()
```
